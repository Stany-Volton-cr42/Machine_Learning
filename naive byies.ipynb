```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sn = pd.read_csv('Social_Network_Ads.csv')
```

```python
sn.head()
```

```python
sn.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
x = sn.iloc[:,[1,2,3]].values
```

```python
y = sn.iloc[:,-1].values
```

```python
x[:,0] = lb.fit_transform(x[:,0])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
# scale our dataset or standrised or numaralized the data
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
modal = GaussianNB()
```

```python
modal.fit(x_train,y_train)
```

```python
y_pred = modal.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
# in confussion matrix we want to know that our error is 5 and 7 
confusion_matrix(y_test,y_pred)
```
